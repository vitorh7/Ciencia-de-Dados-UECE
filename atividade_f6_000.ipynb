```python
# Usando enumerate para exibir os índices e valores da lista
frutas = ["maçã", "banana", "uva"]
for indice, valor in enumerate(frutas):
    print(indice, valor)
```

```python
# Criando uma lista com os números pares de 1 a 10 usando List Comprehension
pares = [x for x in range(1, 11) if x % 2 == 0]
print(pares)
```

```python
# Usando map para multiplicar cada elemento da lista [1, 2, 3, 4] por 3
numeros = [1, 2, 3, 4]
multiplicados = list(map(lambda x: x * 3, numeros))
print(multiplicados)
```

```python
# Usando enumerate e List Comprehension para criar uma lista de tuplas (índice, valor²)
valores = [2, 4, 6]
resultado = [(indice, valor ** 2) for indice, valor in enumerate(valores)]
print(resultado)
```
